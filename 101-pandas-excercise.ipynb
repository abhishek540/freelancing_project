````python
#creating a list , numpy array, dict
import numpy as np
mylist = list("abcdefghijklmnopqrstuvwxyz")
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))

print("List:\t",mylist)
print("Array(numpy):\t",myarr)
print("Dictionary",mydict)

len(mylist)
````

````python
series_list = pd.Series(mylist)
series_np = pd.Series(myarr)
series_dict = pd.Series(mydict)

print(series_list,"\n",series_np,"\n",series_dict)
````

## 6. How to get the items of series A not present in series B?


````python
series1 = pd.Series([1,2,3,4,5])
series2 = pd.Series([4,5,6,7,8])
series1[~series1.isin(series2)]
````

````python
series1.isin(series2)
````

## 7. How to get the items not common to both series A and series B?

````python
a = pd.Series([1, 2, 3, 4, 5])
b = pd.Series([4, 5, 6, 7, 8])

U = pd.Series(np.union1d(a,b))
I = pd.Series(np.intersect1d(a,b))

U[~U.isin(I)]
````

## 8. How to get the minimum, 25th percentile, median, 75th, and max of a numeric series?

````python
a = pd.Series(np.random.normal(10,5,25))

np.percentile(a, q = [0,25,50,75,100])
````
